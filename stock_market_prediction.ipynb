### **Clossing price and SMA indicator on AAPL Stock**

```python
# Load the prices-split-adjusted dataset
prices_split_adjusted = pd.read_csv('prices-split-adjusted.csv')

# Convert the date column to datetime
prices_split_adjusted['date'] = pd.to_datetime(prices_split_adjusted['date'])

# Filter data for a specific stock, e.g., 'AAPL'
stock_data = prices_split_adjusted[prices_split_adjusted['symbol'] == 'AAPL'].copy()
stock_data.set_index('date', inplace=True)

# Calculate SMA (Simple Moving Average)
stock_data['SMA_50'] = stock_data['close'].rolling(window=50).mean()
stock_data['SMA_200'] = stock_data['close'].rolling(window=200).mean()

# Calculate MACD (Moving Average Convergence Divergence)
exp1 = stock_data['close'].ewm(span=12, adjust=False).mean()
exp2 = stock_data['close'].ewm(span=26, adjust=False).mean()
stock_data['MACD'] = exp1 - exp2
stock_data['Signal_Line'] = stock_data['MACD'].ewm(span=9, adjust=False).mean()

# Calculate RSI (Relative Strength Index)
delta = stock_data['close'].diff(1)
gain = delta.where(delta > 0, 0)
loss = -delta.where(delta < 0, 0)
avg_gain = gain.rolling(window=14).mean()
avg_loss = loss.rolling(window=14).mean()
rs = avg_gain / avg_loss
stock_data['RSI'] = 100 - (100 / (1 + rs))

# Drop NaN values
stock_data.dropna(inplace=True)
```

```python
stock_data
```

```python
# Plot the closing price along with SMA
plt.figure(figsize=(14, 7))
plt.plot(stock_data['close'], label='Close Price')
plt.plot(stock_data['SMA_50'], label='50-Day SMA')
plt.plot(stock_data['SMA_200'], label='200-Day SMA')
plt.title('AAPL Close Price and SMA')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

### **Volume Chart on GOOGL Stock**

```python
# Load the prices-split-adjusted dataset
prices_split_adjusted = pd.read_csv('prices-split-adjusted.csv')

# Convert the date column to datetime
prices_split_adjusted['date'] = pd.to_datetime(prices_split_adjusted['date'])

# Filter data for a specific stock, e.g., 'GOOGL'
stock_data = prices_split_adjusted[prices_split_adjusted['symbol'] == 'GOOGL'].copy()
stock_data.set_index('date', inplace=True)

# Drop NaN values
stock_data.dropna(inplace=True)

# Plot the Volume Chart along with Close Price
fig, ax1 = plt.subplots(figsize=(14, 7))

# Plot the close price on the primary y-axis
ax1.plot(stock_data.index, stock_data['close'], label='Close Price', color='tab:blue')
ax1.set_xlabel('Date')
ax1.set_ylabel('Close Price', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Plot the volume on the secondary y-axis
ax2 = ax1.twinx()
ax2.bar(stock_data.index, stock_data['volume'], label='Volume', color='tab:gray', alpha=0.3)
ax2.set_ylabel('Volume', color='tab:gray')
ax2.tick_params(axis='y', labelcolor='tab:gray')

fig.suptitle('GOOGL Close Price and Volume')
fig.tight_layout()
plt.show()
```

### **Histogram of GOOGl Daily Return**

```python
# Load the prices-split-adjusted dataset
prices_split_adjusted = pd.read_csv('prices-split-adjusted.csv')

# Convert the date column to datetime
prices_split_adjusted['date'] = pd.to_datetime(prices_split_adjusted['date'])

# Filter data for a specific stock, e.g., 'GOOGL'
stock_data = prices_split_adjusted[prices_split_adjusted['symbol'] == 'GOOGL'].copy()
stock_data.set_index('date', inplace=True)

# Calculate daily returns
stock_data['daily_return'] = stock_data['close'].pct_change()

# Drop NaN values
stock_data.dropna(inplace=True)

# Plot the histogram of daily returns
plt.figure(figsize=(10, 6))
plt.hist(stock_data['daily_return'], bins=50, alpha=0.75, color='blue', edgecolor='black')
plt.title('Histogram of GOOGL Daily Returns')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
